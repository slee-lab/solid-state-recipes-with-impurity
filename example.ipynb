```python
import pickle
import gzip
with gzip.open("SS_rxns_80823.pkl.gz", "rb") as f:
    SS_rxns = pickle.load(f)
len(SS_rxns),SS_rxns[0].keys()
```

```python
SS_phaseimpure,SS_phasepure = [], []
for rxn in SS_rxns:
    if rxn['impurity_phase']:
        SS_phaseimpure.append(rxn)
    else:
        SS_phasepure.append(rxn)
len(SS_phaseimpure), len(SS_phasepure), len(SS_rxns)
```

```python
from pprint import pprint
for key in SS_rxns[32].keys():
    print(key)
    pprint(SS_rxns[32][key])
    print()
```

```python
def get_formulas(rxn,key):
    """
    key: 'target', 'precursors', or 'impurity_phase'
    """
    return [mat['material_formula'] for mat in rxn[key]]

print(SS_phaseimpure[0]['DOI'])
for key in ['target', 'precursors','impurity_phase']:
    print(key, get_formulas(SS_phaseimpure[32],key))

print(SS_phasepure[0]['DOI'])
for key in ['target', 'precursors','impurity_phase']:
    print(key, get_formulas(SS_phasepure[32],key))
```
